```python
import pandas as pd
import statsmodels.api as sm
df = pd.read_csv("dataset/train.csv", sep=',', parse_dates=[2])

df.head(10)
```

```python
from matplotlib import pyplot
from pandas.plotting import autocorrelation_plot

store_one_sales = df.loc[df['Store'] == 1.0].sort_values(by='Date')

store_one_sales
```

```python
autocorrelation_plot(store_one_sales['Sales'].head(50))
pyplot.show()
```

```python
monday_values = store_one_sales.loc[store_one_sales['DayOfWeek'] == 1]
autocorrelation_plot(monday_values['Sales'].head(10))
pyplot.show()
monday_values
```
