```python
import pandas as pd
import statsmodels.api as sm
df = pd.read_csv("dataset/train.csv", sep=',', parse_dates=[2])

df.head(10)
```

```python
df.corr()
```

```python
# LR input
featureColumns = ['Promo', 'Open', 'DayOfWeek', 'SchoolHoliday']
targetColumn = 'Sales'

inputData = pd.DataFrame(df, columns=featureColumns)
inputData = sm.add_constant(inputData, has_constant='add')
targetData = pd.DataFrame(df, columns=[targetColumn])

# Run LR
model = sm.OLS(targetData, inputData).fit()
model.summary()
```
